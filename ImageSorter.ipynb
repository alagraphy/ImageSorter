```python
#images in unsorted will be copied to sorted
f = glob.glob('../unsorted/*.jpg')
l1 = []

for i in f:
    c = ColorThief(i).get_palette(color_count=4, quality=2)
    c1 = []
    for cc in c:
        c1.append('#'+''.join(map(chr, cc)).encode('hex'))
        
    c1.append( i[12:] )
    l1.append(c1)
    
```
